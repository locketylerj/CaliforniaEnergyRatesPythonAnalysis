```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import date
from datetime import timedelta
from scipy import stats
```

```python
file= 'Henry_Hub_Natural_Gas_Spot_Price.csv'
```

```python
df = pd.read_csv(file)
```

```python
df = df.rename(index=str, columns={"Henry Hub Natural Gas Spot Price Dollars per Million Btu":"HH Spot $/MMBtu"})
df = df.reindex(index=df.index[::-1])
df = df.dropna()
df
```

```python
prices = df["HH Spot $/MMBtu"]
ax = df.plot(x='Month', y='HH Spot $/MMBtu', xticks=(np.arange(0,258,35)), grid=True, title="HH Natural Gas Prices")
fig = ax.get_figure()
ax.figsize=(15,10)
ax.set_ylabel("$/MMBtu", fontsize=10)
z = np.polyfit(np.arange(0,258,1), prices, 1)
p = np.poly1d(z)
plt.plot((0,258,1),p((0,258,1)), "r--")
fig.savefig('HH Natural Gas.png', dpi=200)
```
