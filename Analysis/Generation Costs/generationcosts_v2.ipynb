```python
# Import Dependencies 
import pandas as pd
from pprint import pprint
import matplotlib.pyplot as plt
```

```python
# Read the CSV into the Pandas DataFrame
file = "generationcosts.csv"
df = pd.read_csv(file)
```

```python
# Pretty print the list of all columns for reference
pprint(df.columns)
df
```

```python
# Using loc, grab the data contained within the Technology / Gen Type for each energy source and set into DataFrame
df_nuclear = df.loc[df["Technology"]== "Nuclear"]
df_hydro = df.loc[df["Technology"]== "Hydroelectric"]
df_ng = df.loc[df["Technology"] == "Combined Cycle"]
df_ngPeaker = df.loc[df["Technology"] == "Combustion Turbine"]
df_solar = df.loc[df["Technology"] == "Photovoltaic"]
df_coal = df.loc[df["Gen Type"] == "Coal"]
df_wind = df.loc[df["Technology"] == "Onshore"]
```

```python
# Create boxplots that show the cost of energy per resource
# Boxplot for Nuclear

plt1 = plt.boxplot(df_nuclear["Levelized cost of energy"], patch_artist=True)
plt.title('Nuclear')
plt.xlabel('Levelized Cost of Energy')

for box in plt1['boxes']:
    # change outline color
    box.set(color='gold', linewidth=4)
    # change fill color
    box.set(facecolor = 'lightgreen' )
    # change hatch
    box.set(hatch = '/', linewidth=1)
    

```

```python
plt2 = plt.boxplot(df_hydro["Levelized cost of energy"], patch_artist=True)
plt.title('HydroElectric')
plt.xlabel('Levelized Cost of Energy')


for box in plt2['boxes']:
    # change outline color
    box.set(color='gold', linewidth=4)
    # change fill color
    box.set(facecolor = 'aqua' )
    # change hatch
    box.set(hatch = '/', linewidth=1)
```

```python
# Boxplot for Combined Cycle
plt3 = plt.boxplot(df_ng["Levelized cost of energy"], patch_artist=True)
plt.title('Combined Cycle')
plt.xlabel('Levelized Cost of Energy')

for box in plt3['boxes']:
    # change outline color
    box.set(color='red', linewidth=4)
    # change fill color
    box.set(facecolor = 'lightpink' )
    # change hatch
    box.set(hatch = '/', linewidth=1)
```

```python
# Boxplot for Combustion Turbine
plt4 = plt.boxplot(df_ngPeaker["Levelized cost of energy"], patch_artist=True)
plt.title('Combustion Turbine')
plt.xlabel('Levelized Cost of Energy')

for box in plt4['boxes']:
    # change outline color
    box.set(color='orange', linewidth=4)
    # change fill color
    box.set(facecolor = 'silver' )
    # change hatch
    box.set(hatch = '/', linewidth=1)
```
